```python
import numpy
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
from glob import glob
```

```python
images = glob('/densepose_mydata/test_imgs/*.jpg')
```

```python
all_data = []
for im in images:
    im_dict = {}
    im_dict['im'] = im
    im_dict['IUV'] = os.path.join('/densepose_mydata/infer_out', os.path.basename(im).split('.')[0]+'_IUV.png')
    im_dict['INDS'] = os.path.join('/densepose_mydata/infer_out', os.path.basename(im).split('.')[0]+'_INDS.png')
    all_data.append(im_dict)
```

```python
n = 8

im = cv2.imread(all_data[n]['im'])
IUV = cv2.imread(all_data[n]['IUV'])
INDS = cv2.imread(all_data[n]['INDS'], 0)
print(INDS)
```

Visualize the I, U and V images.

```python
fig = plt.figure(figsize=[15,15])
plt.imshow(   np.hstack((IUV[:,:,0]/24. ,IUV[:,:,1]/256. ,IUV[:,:,2]/256.))  )
plt.title('I, U and V images.')
plt.axis('off') ; plt.show()
```

Visualize the isocontours of the UV fields.

```python
fig = plt.figure(figsize=[12,12])
plt.imshow( im[:,:,::-1] )
plt.contour( IUV[:,:,1]/256.,10, linewidths = 1 )
plt.contour( IUV[:,:,2]/256.,10, linewidths = 1 )
plt.axis('off') ; plt.show()
```

Visualize the human-body FG mask indices.

```python
fig = plt.figure(figsize=[12,12])
plt.imshow( im[:,:,::-1] )
plt.contour( INDS, linewidths = 4 )
plt.axis('off') ; plt.show()
```
